# Importing the libraries

```python
import numpy as np  #Numpy is widely used library for handling number
import matplotlib.pyplot as plt #Provides a library for data visualization for plotting graphs
import pandas as pd  #Used for data manipulation
```

# Importing the dataset

```python
dataset = pd.read_csv("/content/drive/MyDrive/Salary_Data_simple lin.csv") #Pull the Data.csv file from CSV into a variable.
X = dataset.iloc[:,:-1].values #Indexing all the values till the last row.
y = dataset.iloc[:,-1].values #Indexing all the values till the last column.
```

# Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split #Splitting data into random train and test basically converting the data into a one-liner
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state= 1) #Assign the test size at 0.3 so 70% of the data will be used for training and 30% will be used for testing.Random state is used for deciding the splitting of a given data by initializing the internal random number 
```

# Training the Simple Linear Regression model on the training set

```python
from sklearn.linear_model import LinearRegression #Import the linear regression packages for the purposes of Linear approximations.
regressor = LinearRegression() #This function enables fitting linear models to the data frames
regressor.fit(X_train, y_train) #The .fit() method helps implementing fitting line to the data.
```

# Predicting the Test set results

```python
y_pred = regressor.predict(X_test) # Storing the predicted values for all the values of x_test.
```

# Visualizing the Training set results

```python
plt.scatter(X_train,y_train,color = 'blue') #The scatter() function plots one dot for each observation according to the input values.
plt.plot(X_train, regressor.predict(X_train),color = 'red') #The plot() function draws a line by using points.
plt.title('Salary vs Experience (Training set)') #Used for assigning title to the graph
plt.xlabel('Years of Experience') #Used for labelling x axis.
plt.ylabel('Salary') #Used for labelling y axis.
plt.show() #Used for displaying the graph
```

#Visualizing the Test set results

```python
plt.scatter(X_test,y_test,color = 'red') #The scatter() function plots one dot for each observation according to the input values.
plt.plot(X_train, regressor.predict(X_train),color = 'blue') #The plot() function draws a line by using points.
plt.title('Salary vs Experience (Training set)')  #Used for assigning title to the graph
plt.xlabel('Years of Experience') #Used for labelling x axis.
plt.ylabel('Salary') #Used for labelling y axis.
plt.show()  #Used for displaying the graph
```
