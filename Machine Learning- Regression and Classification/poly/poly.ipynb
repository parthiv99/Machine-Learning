#Importing The Libraries

```python
import numpy as np  #Numpy is widely used library for handling number
import matplotlib.pyplot as plt #Provides a library for data visualization for plotting graphs
import pandas as pd  #Used for data manipulation
```

#Importing the dataset

```python
dataset = pd.read_csv("/content/drive/MyDrive/Position_Salaries_poly.csv") #Pull the Data.csv file from CSV into a variable.
x = dataset.iloc[:,1:-1].values #Indexing all the values till the last row.
y = dataset.iloc[:,-1].values #Indexing all the values till the last column.
```

#Training the Linear Regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression #Import the linear regression packages for the purposes of Linear approximations.
lin_reg = LinearRegression() #This function enables fitting linear models to the data frames
lin_reg.fit(x,y) #The .fit() method helps implementing fitting line to the data.
```

#Training the Polynomial Regression model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures #Libraries needed for using differnet polynomial combinations that depends on the degree specified.
poly_reg = PolynomialFeatures(degree = 6) #Assigning a degree of 3, so restricting the number of features to 4.
x_poly = poly_reg.fit_transform(x) #Fitting the data and than transformation the data of x.
lin_reg_2 = LinearRegression() #Performing linear regression on the polynomial combination.
lin_reg_2.fit(x_poly,y) #Fitting the lin_reg_2 using x_poly.
```

#Visualising the Linear Regression Results

```python
plt.scatter(x,y,color = 'blue') #The scatter() function plots one dot for each observation according to the input values.
plt.plot(x,lin_reg.predict(x) ,color = 'red') plt.plot(x,lin_reg_2.predict(poly_reg.fit_transform(x)) ,color = 'red') #The plot() function draws a line by using points.
plt.title('Truth or Bluff (Linear Regression)') #Used for assigning title to the graph
plt.xlabel('Position Level') #Used for labelling x axis.
plt.ylabel('Salary') #Used for labelling y axis.
plt.show() #Used for displaying the graph
```

#Visualising the Polynomial Regression Results

We see the importance of using Polynomial Regression with the graph below that we are able to visualize a high resolution and a more accurate curve reading.

```python
plt.scatter(x,y,color = 'blue') #The scatter() function plots one dot for each observation according to the input values.
plt.plot(x,lin_reg_2.predict(poly_reg.fit_transform(x)) ,color = 'red') #The plot() function draws a line by using points.
plt.title('Truth or Bluff (Linear Regression)') #Used for assigning title to the graph
plt.xlabel('Position Level') #Used for labelling x axis.
plt.ylabel('Salary') #Used for labelling y axis.
plt.show() #Used for displaying the graph
```

#Predicting a new result with Linear Regression

```python
lin_reg.predict([[5]]) #The predict function helps in predicting a new result with Linear Regression.
```

#Predicting a new result with Polynomial Regression

```python
lin_reg_2.predict(poly_reg.fit_transform([[5]])) #The fit and transform method allow us to predict a new result with Polynomial Regression.
```
