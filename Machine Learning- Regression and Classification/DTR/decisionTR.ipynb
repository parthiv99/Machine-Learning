#Importing the libraries

```python
import numpy as np  #Numpy is widely used library for handling number
import matplotlib.pyplot as plt #Provides a library for data visualization for plotting graphs
import pandas as pd  #Used for data manipulation
```

#Importing the dataset

```python
dataset = pd.read_csv("/content/drive/MyDrive/Data_regression_project.csv") #Pull the Position_Salaries_poly-1.csv file from CSV into a variable.
x_values = dataset.iloc[:,1:2].values #Indexing all the values till the specific row.
y = dataset.iloc[:, 2].values #Indexing all the values till the specific column.
```

#Training the DTR model on the whole dataset

```python
from sklearn.tree import DecisionTreeRegressor #Importing Decision Tree regression regressor.
regressor = DecisionTreeRegressor(random_state = 0) #Creating a Decision Tree Regressor object with a random state at 0.  
regressor.fit(x_values, y) #Fitting the regressor with x_values and y from the given dataset.
```

#Predicting a new result

```python
y_pred = regressor.predict([[3200]]) #Predicing a new value to test the output by assigning a value of 3200
print(y_pred)  #Printing the predicted value y_pred.
```

#Visualising the DTR results (High resolution)

```python
X_grid = np.arange(min(x_values), max(x_values), 0.03)  #Arranging to assign limited range of data and with a difference of 0.03 between two consecutive values.
X_grid = X_grid.reshape((len(X_grid), 1)) #Reshaping the data into a specific form to get the desired output.
plt.scatter(x_values, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue') 
plt.title('Profit to Production Cost (Decision Tree Regression)') 
plt.xlabel('Production Cost')
plt.ylabel('Profit')
plt.show()
```
