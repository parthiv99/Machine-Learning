#Importing the libraries




```python
import numpy as np  #Numpy is widely used library for handling number
import matplotlib.pyplot as plt #Provides a library for data visualization for plotting graphs
import pandas as pd  #Used for data manipulation
```

#Importing the dataset.

```python
dataset = pd.read_csv("/content/drive/MyDrive/Data_regression_project.csv") #Pull the Position_Salaries_poly-1.csv file from CSV into a variable.
x_values = dataset.iloc[:, 1:2].values #Indexing all the values till a specific row.
y = dataset.iloc[:, 2].values #Indexing all the values till a specific column.
```

#Printing the x_values

```python
print(x_values) #Printing x_values data value. 
```

#Training the RFR model on the whole dataset

```python
from sklearn.ensemble import RandomForestRegressor #Importing the regressor for fitting RFR to the given dataset/
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0) #Create regressor object that fits an amount of decision trees on multiple sub-samples in the dataset.n_estimaters is used to assign the nummber of trees in the forest. random_state is responsible for controlling the randomness.  
regressor.fit(x_values, y) #Fitting  the regression with x_values and y data in the given dataset.
```

#Predicting a new result

```python
Y_pred = regressor.predict(np.array([6.5]).reshape(1, 1))  #testing the predictedoutput by changing values.
```

#Visualising the Random Forest Regression results. (High Resolution)

```python
X_grid = np.arange(min(x_values), max(x_values), 0.01) #Arranging to assign limited range of data                
X_grid = X_grid.reshape((len(X_grid), 1)) #Reshaping the data into a specific form to get the desired output.
plt.scatter(x_values, y, color = 'blue')  #Scatter plot of the given data.
plt.plot(X_grid, regressor.predict(X_grid),color = 'red') #Predicting and graphing using the data. 
plt.title('Random Forest Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
