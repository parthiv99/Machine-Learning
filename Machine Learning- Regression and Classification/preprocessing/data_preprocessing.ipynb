# Importing the libraries

```python
import numpy as np  #Numpy is widely used library for handling number
import matplotlib.pyplot as plt #Provides a library for data visualization for plotting graphs
import pandas as pd  #Used for data manipulation
```

# Importing the dataset

```python
dataset = pd.read_csv("/content/drive/MyDrive/Data.csv") #Pull the Data.csv file from CSV into a variable.
x_values = dataset.iloc[:,:-1].values #Indexing all the values till the last row.
y_values = dataset.iloc[:,-1].values #Indexing all the values till the last column.
```

```python
print(x_values) #Printing all the X values
```

```python
print(y_values) #Printing all the y values
```

# Taking care of missing data

```python
from sklearn.impute import SimpleImputer #SimpleImputer is used for imputing missing values
imputer = SimpleImputer(missing_values=np.nan, strategy = 'median') #This implies replacing missing values with different strategies like mean and median. 
imputer.fit(x_values[:,1:3]) #The : refers to selecting every row in the data set, 1:3 indicates selecting first 3 columns. 
x_values[:,1:3] = imputer.transform(x_values[:,1:3])
```

```python
print(x_values)
```

# Encoding the Categorical data
## Encoding independent variable

```python
from sklearn.compose import ColumnTransformer #Used for applying transformers to the columns. 
from sklearn.preprocessing import OneHotEncoder #OneHotEncoder is used to create binary column for every value and return a sparse matrix.
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])], remainder='passthrough' ) #Creating  a column Transformer.  
x_values = np.array(ct.fit_transform(x_values)) #Next, we fit and transform the created column transformaer to a label encode.
```

```python
print(x_values)
```

# Encoding Dependent Variable

```python
from sklearn.preprocessing import LabelEncoder #Encode target values and the variable y is the target value.
le = LabelEncoder() #This function is helpful for encoding labels to specific classes.
y_values = le.fit_transform(y_values) #Indicating a use of a combination of fit and transform method.
```

```python
print(y_values)
```

# Splitting the dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split #Splitting data into random train and test basically converting the data into a one-liner
x_train, x_test, y_train,y_test = train_test_split(x_values,y_values,test_size = 0.2, random_state= 1) #Assign the test size at 0.2 so 80% of the data will be used for training and 20% will be used for testing.Random state is used for deciding the splitting of a given data by initializing the internal random number 
```

```python
print(x_train) #Printing all the independent variables that will be used for training purposes for x.
```

```python
print(x_test) #Printing the x_test that shows 20% that will be used for testing the accuracy.
```

```python
print(y_train) #Printing the dependent variable
```

```python
print(y_test) #This prints the values that will be used for testing the accuracy.
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler() #This function helps standardize features by removing the mean and scaling to unit variance to assign a standard format to the data values.
x_train[:,3:] = sc.fit_transform(x_train[:,3:]) #Use a combination of fit and transform method for training.
x_test[:,3:] = sc.transform(x_test[:,3:]) #testing the x values for unbiased evaluation and it only uses transform method because we cannot use fit method.
```
