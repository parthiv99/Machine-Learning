#Importing the libraries

```python
import numpy as np  #Numpy is widely used library for handling number
import matplotlib.pyplot as plt #Provides a library for data visualization for plotting graphs
import pandas as pd  #Used for data manipulation
```

#Importing the Dataset

```python
dataset = pd.read_csv("/content/drive/MyDrive/Startups_multi.csv") #Pull the Data.csv file from CSV into a variable.
x_values = dataset.iloc[:,:-1].values #Indexing all the values till the last row.
y = dataset.iloc[:,-1].values #Indexing all the values till the last column.
```

```python
print(x_values)
```

#Encoding Categorical data

```python
from sklearn.compose import ColumnTransformer #Used for applying transformers to the columns. 
from sklearn.preprocessing import OneHotEncoder #OneHotEncoder is used to create binary column for every value and return a sparse matrix.
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[3])], remainder='passthrough' ) #Creating  a column Transformer.  
x_values = np.array(ct.fit_transform(x_values)) #Next, we fit and transform the created column transformaer to a label encode.
```

```python
print(x_values)
```

#Splitting the dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split #Splitting data into random train and test basically converting the data into a one-liner
X_train,X_test,y_train,y_test = train_test_split(x_values,y,test_size = 0.3,random_state= 1) #Assign the test size at 0.3 so 70% of the data will be used for training and 30% will be used for testing.Random state is used for deciding the splitting of a given data by initializing the internal random number 
```

#Training the Multiple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression #Import the linear regression packages for the purposes of Linear approximations.
regressor = LinearRegression() #This function enables fitting linear models to the data frames
regressor.fit(X_train, y_train) #The .fit() method helps implementing fitting line to the data.
```

#Predicting the Test set results

```python
y_pred = regressor.predict(X_test) # Storing the predicted values for all the values of x_test.
np.set_printoptions(precision = 3) # set_printoptions function is used for setting the direction of floating point numbers and array can be displayed.
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1)) #Printing the concatenated form after reshaping the arrays of y_test and y_pred.
```
